```r
library('CrossTalkeR')
library('igraph')
library("factoextra")
library("ggplot2")
library("ggrepel")
# the method always consider the first path as control: the multiple control case will be handle soon
paths <- c('CTR' = system.file("extdata",
                               "ctr_nils_bm_human_newformat.csv",
                               package = "CrossTalkeR"),
           'EXP' = system.file("extdata",
                               "exp_nils_bm_human_newformat.csv",
                               package = "CrossTalkeR"))
```

```r
data <- read.csv(paths[[1]])
```

```r
head(data)
```

```r
# Setting some interesting genes      
genes <- c('TGFB1') 
# It can be also 
genes <- NULL 
```

```r
output =  system.file("extdata", package = "CrossTalkeR") ### Please replace it accordingly
data <- generate_report(paths,
                        genes,
                        out_path=paste0(output,'/'),
                        threshold=0,
                        out_file = 'vignettes_example.html',
                        output_fmt = "html_document",
                        report = FALSE)
```

## General Overview CCI

```r
options(repr.plot.width=10, repr.plot.height=10)
plot_cci(graph = data@graphs$CTR, ## Selecting the Cell Cell Network
        colors = data@colors[V(data@graphs$CTR)$name], ## Setting the color Pallet
        plt_name = 'Example 1', ## Plot title
        coords = data@coords[V(data@graphs$CTR)$name,], ## Nodes Coordinates
        emax = NULL,
        leg = TRUE, ## Setting Legend
        low = 0,
        high = 0,
        ignore_alpha = FALSE,
        log = FALSE,
        efactor = 8,
        vfactor = 12,
        pg=data@rankings$CTR$Pagerank[V(data@graphs$CTR)$name])
```

## Node ranking methods higher scored candidates

```r
head(data@rankings$EXP_x_CTR_ggi)
```

```r
x <- max(abs(data@pca$EXP_x_CTR_ggi$x[,1]))
y <- max(abs(data@pca$EXP_x_CTR_ggi$x[,2]))
z_x <- data@pca$EXP_x_CTR_ggi$x[,1]
z_y <- data@pca$EXP_x_CTR_ggi$x[,2]
ver_zx <- ifelse(abs(z_x)>=(2*data@pca$EXP_x_CTR_ggi$sdev[1]),1,0)
ver_zy <- ifelse(abs(z_y)>=(2*data@pca$EXP_x_CTR_ggi$sdev[2]),1,0)
## Setting max.overlaps=25 you can increase or decrease the labels
## All the labels needs to be at 2 stdev in both PCA axis
fviz_pca_biplot(data@pca$EXP_x_CTR_ggi,
                    axes = c(1,2),
                    pointshape = 21, pointsize = 0.5,labelsize = 10,
                    repel = FALSE,max.overlaps=200,label='var')+
                    geom_label_repel(aes(label=ifelse((ver_zx | ver_zy),rownames(data@pca$EXP_x_CTR_ggi$x),NA)),size = 5,max.overlaps=25)+                     
                    xlim(-x, x)+
                    ylim(-y, y)+
                    ggtitle('PCA example')+
                    theme(text = element_text(size = 7.5),
                          axis.title = element_text(size = 7.5),
                          axis.text = element_text(size = 7.5))
```
