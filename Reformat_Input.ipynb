```python
import pandas as pd
```

```python
data = pd.read_csv('./inst/extdata/ctr_nils_bm_human.csv')
newdata =pd.DataFrame()
newdata['source'] = data['Ligand.Cluster']
newdata['target'] = data['Receptor.Cluster']
newdata['gene_A'] = data['Ligand']
newdata['gene_B'] = data['Receptor']
newdata['type_gene_A'] = 'Ligand'
newdata['type_gene_B']= 'Receptor'
newdata['MeanLR']=data['MeanLR']
newdata.to_csv('./inst/extdata/ctr_nils_bm_human_newformat.csv')
```
