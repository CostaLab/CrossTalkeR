```r
require(liana)
require(tibble)
require(purrr)
require(Seurat)
require(reticulate)
require(tidyverse)
require(CrossTalkeR)
devtools::load_all("../../CrossTalkeR/")
liana_path <- system.file(package = "liana")
testdata <-
  readRDS(file.path(liana_path , "testdata", "input", "testdata.rds"))
testdata %>% glimpse()
```

```r
liana_test <- liana_wrap(testdata)
```

```r

 data <- liana_test$cellphonedb %>% 
    filter(pvalue<=0.05) %>%
    mutate(type_gene_A="Ligand") %>%
    mutate(type_gene_B="Receptor") 
input <- list()
input[['test']] <-data
data <- generate_report(input,
                genes=NULL,
                out_path='~/',
                threshold=0,
                out_file = 'test.html',
                output_fmt = "html_document",
                report = F,
                sel_columns = c('source',
                                'target',
                                'ligand',
                                'receptor',
                                'type_gene_A','type_gene_B','lr.mean')) 
```
